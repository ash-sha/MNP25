```python
from importlib.metadata import Distribution
import pandas as pd
```

```python
master_data = pd.read_excel("/Users/aswathshakthi/PycharmProjects/MNP25/master_data/Master_Records_1-03-2025 WIP.xlsx")
master_data = master_data[master_data["Quantity"]!=0].reset_index(drop=True)
master_data = master_data.drop(columns=["Sequence List"])
master_data
```

```python
master_data.isna().sum()
```

```python
master_data.loc[master_data['Beneficiary Type'] != 'District', 'Names'] = master_data['App. No.'] + ' - ' + master_data['Name']
master_data.loc[master_data['Beneficiary Type'] == 'District', 'Names'] = master_data['Name']
master_data = master_data.drop(columns=["Name","App. No."])
cols = list(master_data.columns)
cols.insert(1, cols.pop(cols.index('Names')))  # Move 'Concatenated' to second position
df = master_data[cols]
df
```

```python
df["Total Value"].sum()
```

```python
grouped = df.groupby(["Names","Article Name","ITEM TYPE","Beneficiary Type","Article Category"],as_index=False).agg({"Quantity": "sum", "Total Value":"sum"})

grouped
```

```python
grouped[grouped[["Names","Article Name"]].duplicated(keep=False)]
```

```python
distribution_ls = pd.read_excel("/Users/aswathshakthi/PycharmProjects/MNP25/master_data/District wise article waiting hall list.xlsx",sheet_name="Sheet-new").drop(columns=["Sl.No."])
distribution_ls = distribution_ls.rename(columns={"Districts ":"Names","Article list":"Article Name","Total Quantity":"Quantity","Total Values":"Total Value"})
distribution_ls
```

```python
step2 = grouped.merge(distribution_ls, on=["Names","Article Name","Quantity","Total Value"],how="left")
step2
```
